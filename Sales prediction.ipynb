```python
# import required packages for preprocessing data,model making and visualization

import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.simplefilter("ignore")
```

```python
# loading the given dataset for sales
data = pd.read_csv('C:\\Users\\tamil\\Downloads\\advertising.csv')
```

```python
# calling head() to display few data from the dataset loaded
data.head()
```

```python
# checking for null values in the dataset
data.isna().sum()
```

```python
# spliting the dataset so can use in variables separately
# lets split the data into X and Y 
# X - independent variables , Y - dependent variable
X = data.drop(columns=['Sales'])
Y = data['Sales']
```

```python
sb.histplot(x='TV',data=data)
plt.title("TV advirtisement")
```

```python
sb.histplot(x='Radio',data=data)
plt.title("Radio advirtisement")
```

```python
sb.histplot(x='Newspaper',data=data)
plt.title("Newspaper advirtisement")
```

```python
plt.figure(figsize=(14,8))
sb.heatmap(data.corr(),annot=True,cmap='twilight')
plt.title('Heatmap for data representation')
```

```python
# spliting the data at a ratio of 70:30 for training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
print("shape of X-train : ",X_train.shape)
print("shape of X-test : ",X_test.shape)
print("shape of Y-train : ",Y_train.shape)
print("shape of Y-test : ",Y_test.shape)
```

```python
# creating a model of multiple regression
model = LinearRegression()

# fitting the model with the data
model.fit(X_train,Y_train)

# predicting using the test data
pred=model.predict(X_test)
```

```python
#predicting the output by random inputs representing independent variables
print("prediction 1 :",model.predict([[240.3,34.2,65.9]]))
print("prediction 2 :",model.predict([[17.2,45.9,69.3]]))
```

```python
# checking the accuracy of the model we created
from sklearn.metrics import r2_score
print("accuracy_score",r2_score(Y_test,pred))
```
